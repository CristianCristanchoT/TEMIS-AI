```python
import os
import subprocess
from subprocess import Popen, PIPE
import glob
from tqdm import tqdm 
import pandas as pd
import re
```

```python
data = pd.DataFrame(columns=['INDEX','TEXT'])
```

```python
source_folder = 'C:/Users/crist/Desktop/TEMIS-AI/1_WebScrapping/v2/DATAFRAME_CREATION/TXT'
files_path = glob.glob(source_folder + '/*', 
                   recursive = True)
files = []

for file_path in tqdm(files_path):
    file = file_path.replace('C:/Users/crist/Desktop/TEMIS-AI/1_WebScrapping/v2/DATAFRAME_CREATION/TXT\\','')
    file = file.replace('.txt','')
    files.append(file)
    
```

```python
file_name = files[9399]

file_name = 'C-037-03'
print(file_name)
with open( 'TXT/' + file_name + '.txt', 'r', errors='ignore') as file:
    text = file.read()
    text = re.sub('(--+)', '', text)
print(text)
```

```python
with open( 'TXT/' + file_name + '.txt', 'r', encoding="utf8",errors='ignore') as file:
    text = file.read()
    text = re.sub('(--+)', '', text)
print(text)
```

```python
for file in tqdm(files):
    
    text = ''
    
    try:
        
        txt_file = 'TXT/' + str(file) + '.txt'
        
        with open( txt_file, 'r') as doc_file:
            text = doc_file.read()
            text = re.sub('(--+)', '', text)
        
        new_row = {'INDEX':str(file),'TEXT':text}
        data = data.append(new_row, ignore_index=True)
    
    except: 
        
        try:
            
            txt_file = 'TXT/' + str(file) + '.txt'

            with open( txt_file, 'r', encoding="utf8", errors='ignore') as doc_file:
                text = doc_file.read()
                text = re.sub('(--+)', '', text)

            new_row = {'INDEX':str(file),'TEXT':text}
            data = data.append(new_row, ignore_index=True)
        
        
        except Exception as e:

            print("Error!: " + str(e))
            print('Problema en: '+ file)
        
            
```
